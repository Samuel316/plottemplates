```python
import matplotlib
import matplotlib.pyplot as plt

import numpy as np

plt.style.use('fivethirtyeight')
plt.style.use('seaborn-poster')
plt.style.use('seaborn-colorblind')
```

```python
x = np.linspace(0, 2, 10)
y1 = np.sin(x ** 2)
y2 = 1-y1
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(2,1)

axt = fig.add_subplot(gs[0, :])
axb = fig.add_subplot(gs[1, :], sharex=axt)

axt.scatter(x, y1)
axt.plot(x, y1, linestyle="--")
axt.scatter(x, y2)
axt.plot(x, y2, linestyle="--")


y_error = (y1+1)*0.8
axb.scatter(x, y1, marker="x", c="C0")
axb.errorbar(x, y1, yerr=y_error, linestyle="None", marker="x", capsize=20, elinewidth=3, c="C0")
axb.fill_between(x, y1+y_error, y1-y_error, alpha=0.4, color="C0")

y_error = (y2+1)*0.8
axb.scatter(x, y2, marker="+", c="C1")
axb.errorbar(x, y2, yerr=y_error, linestyle="None" , marker="x", capsize=2, c="C1")
axb.fill_between(x, y2+y_error, y2-y_error, alpha=0.2, color="C1")
```
