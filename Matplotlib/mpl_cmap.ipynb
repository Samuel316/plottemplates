```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-poster')
plt.style.use('seaborn-colorblind')
```

```python
z = np.random.uniform(0,1,(8,8))
print(z)
```

```python
fig = plt.figure(constrained_layout=True, figsize=(12, 6))
gs = fig.add_gridspec(2,3)

axtl = fig.add_subplot(gs[0, 0])
axtm = fig.add_subplot(gs[0, 1])
axtr = fig.add_subplot(gs[0, 2])

axbl = fig.add_subplot(gs[1, 0])
axbm = fig.add_subplot(gs[1, 1])
axbr = fig.add_subplot(gs[1, 2])

fig.colorbar(axtl.imshow(z, cmap='RdBu_r'), ax=axtl)
fig.colorbar(axtm.imshow(z, cmap="PuOr"), ax=axtm)
fig.colorbar(axtr.imshow(z), ax=axtr)

fig.colorbar(axbl.pcolormesh(z, cmap='RdBu_r', ), ax=axbl)
fig.colorbar(axbm.pcolormesh(z, cmap="PuOr"), ax=axbm)
fig.colorbar(axbr.pcolormesh(z), ax=axbr)
```
