```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

plt.style.use('fivethirtyeight')
plt.style.use('seaborn-poster')
plt.style.use('seaborn-colorblind')
```

```python
x = np.linspace(0, 9.5)
y = np.sin(x)
z = np.sin(y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z)
ax.plot(x, y)
ax.plot(x, -y)
```
